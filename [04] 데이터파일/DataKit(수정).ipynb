## 데이터 불러오기

```python
dtypes={'위도':str,'경도':str,'도로명주소':str,'index':str}
```

```python
cate = pd.read_csv('카테고리용데이터.csv',sep=',',dtype=dtypes)
```

```python
df = cate[cate['도로명주소'].str.contains('대전광역시')]
```

```python
df['API']=""
```

```python
# 교통, 마트편의점, 교육시설, 의료시설, 음식점카페, 문화공간 순서
keywords = {'교통':['SW8'],
            '마트편의점':['MT1','CS2'],
            '교육시설':['PS3','SC4','AC5'],
            '의료시설':['HP8','PM9'],
            '음식점카페':['FD6','CE7'],
            '문화공간':['CT1','PO3']}
category = ['교통','마트편의점','교육시설','의료시설','음식점카페','문화공간']
```

```python
def getApiResult(row):
    final = {}
    for c in category:
        keyword = keywords[c]
        sum_count = 0
        for k in keyword:
            try:
                params = {'category_group_code': k, 'x':row['위도'],
                     'y':row['경도'],'radius':1000,'page':1}

                response = requests.get('https://dapi.kakao.com/v2/local/search/category.json', headers=headers, params=params)
                print(k)
                print(response.json()['meta'])
                count = response.json()['meta']['total_count']

                sum_count += count
            except:
                sum_count = "ERR"
                break

        final[c] = sum_count
    return final
```

```python
df['API'] = df.apply(getApiResult, axis=1)
```
