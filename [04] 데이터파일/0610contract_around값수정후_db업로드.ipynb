```python
dll = {}
def getLatLon(row):
    address = row['address']
    if address in dll:
        pass
    else:
        dll[address]={}
        dll[address]['lat']=row['longitude']
        dll[address]['lon']=row['latitude']
contract.apply(getLatLon,axis=1)
```

```python
def getLat(val):
    if val not in dll:
        return "ERROR"
    return dll[val]['lat']
def getLon(val):
    if val not in dll:
        return "ERROR"
    return dll[val]['lon']
around['latitude']=around['address'].apply(getLat)
around['longitude']=around['address'].apply(getLon)
```
